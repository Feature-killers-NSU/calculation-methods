#Задача: дан интеграл $I_N=e^{-1}\int_{0}^{1}x^Ne^xdx$:
1) Требуется вывести реккурентную формулу для $I_N$.

2) Объяснить почему появляются проблемы при вычсилении формулы при больших $N$.



#1)Вывод реккурентной формулы:
$I_N=e^{-1}\int_{0}^{1}x^Ne^xdx$

Воспользуемся формулой интегрирования по частям:

$e^{-1}\int_{0}^{1}x^Ne^xdx=e^{-1}(x^Ne^x\bigg|_0^1 -\int_{0}^{1}Nx^{N-1}e^xdx)=$

$=e^{-1}(e^1-N\int_{0}^{1}x^{N-1}e^xdx)=1-Ne^{-1}\int_{0}^{1}x^{N-1}e^xdx=$

$=1-NI_{N-1}$

#2) Рассмотрим динамику  изменение  $I_N$  в завсимости от увеличения $N$.

```python
import numpy as np
I=1-np.exp(-1)
print(I)
for n in range(1,21):
  I=1-n*I
  print("I_{}:{}".format(n,I))
```

```python
import sympy as sp
I=sp.Symbol("I")
for n in range(1,21):
  I=1-n*I
  print("I_{}:{}".format(n,I))
```

#Построим график зависимости $I=I(N)$.

```python
import matplotlib.pyplot as plt
N=21
x=[i for i in range(N)]
y=[1-np.exp(-1)]
for i in range(1,N):
  y.append(1-i*y[i-1])

plt.plot(x,y)
```

При пристальном вглядывании можно обнаружить, что $I=I_N=I(N)=k_N*I_0+b_N$.

Более того нетрудно заметить, что $k_N=(-1)^NN!$,так как на каждом шаге n мы умножаем на n предыдущее. $b_n=-nb_{n-1}+1$

```python
N=21
x_1=[i for i in range(N)]
k=[(-1)**i*np.math.factorial(i) for i in x_1]
b=[0]
for i in range(1,N):
  b.append(-i*b[i-1]+1)

plt.plot(x_1,k,label="k_n")
plt.plot(x_1,b,label="b_n")
plt.legend()
```
